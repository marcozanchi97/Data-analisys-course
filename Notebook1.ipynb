#### Some commands for lists

Pyhton provides some useful commands specifically for list. The syntax is: name_list.command_name(input_command1,input_command2...)

```python
# count() : returns the number of times that a specified element appears in the list.

ages = [21, 22, 23, 22, 24]

count_23 = ages.count(23) #this command will not return the list, but a number
print(count_23)
print(ages.count(23))

count_22 = ages.count(22)
print(count_22)
```

```python
# append() : adds an item to the end of the list.

ages = [21, 22, 23, 22, 24]

ages.append(25) #this command will return the list


print(ages)
```

```python
# pop() : removes the item at the given index/position
colors_1 = ['red','blu','yellow']
colors_1.pop(0)
print(colors_1)
```

```python
colors_1 = ['red','blu','yellow']
removed_item = colors_1.pop(1)
print(removed_item)
print(colors_1)
```

### 3) For cycle: how to rapidly access and modify list and array

for is a special command in Python. It allows to create an iterative cycle on elements of a list or on elemets of a numpy array.

The syntax is: 

```python
for iterator_name in list_name:
    #command on the iterator
    
```

In practice in each cycle the iterator_name variable assume the value of an element of the list

```python
list_name = [2637,3727,3828,3829]

for element in list_name:
    print(element)   
```

```python
print(list_name[0])
print(list_name[1])
print(list_name[2])
print(list_name[3])
```

```python
# The TAB or the indentation before the command is extremely important
list_name = [2637,3727,3828,3829]

for element in list_name:

print(element)
```

```python
list_name = [2637,3727,3828,3829]

for element in list_name:
    print(element)   

print('For finished')
```

```python
list_name = [2637,3727,3828,3829]

for element in list_name:
    print(element*2)
```

```python
array = np.random.rand(1000)

for element in array:
    print(element)
```

We can also access each element of our list or array cycling on the index:

```python
list_name = [2637,3727,3828,3829]

# Instead of doing like that
for element in list_name:
    print(element)
    
print('')
    

# We can do
for i in range(len(list_name)):
    print(i)
    print(list_name[i])
```

```python
array = np.random.rand(1000)

for index in range(len(array)):
    print(array[index],index)
```
