```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Data/morti-tumori/hlth_cd_aro_1_Data.csv')
df.drop(labels=['Flag and Footnotes','UNIT','RESID'],axis = 1,inplace = True)
df.rename(columns={
        'TIME': 'Year', 
        'GEO': 'Nation',
        'ICD10': 'Disease',
        'Value': 'Number of deaths'
    }, inplace=True)
df.replace(':',np.nan, inplace=True)
```

```python
df = df[(df.Year<2020) & (df.Year>=2017)]
```

```python
df
```

```python
df['Number of deaths'].isnull().sum()
```

```python
nations = df.Nation.unique()
years = df.Year.unique()
```

```python
for nation in nations:
    df_nation = df[df.Nation == nation]
    for year in years:
        df_year = df_nation[df_nation.Year == year]
        print(nation,year, df_year['Number of deaths'].isnull().sum())
```

```python
df.to_csv('Data_cleaned/EU-tumors.csv')
```
