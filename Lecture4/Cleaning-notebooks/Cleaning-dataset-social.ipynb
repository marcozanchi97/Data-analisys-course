```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Data/mean_income/ilc_di17_1_Data.csv')

df.drop(labels=['Flag and Footnotes','INDIC_IL','UNIT'],axis = 1,inplace = True)
df.rename(columns={
        'TIME': 'Year', 
        'GEO': 'Nation',
        'Value': 'Mean income'
    }, inplace=True)
df.replace(':',np.nan, inplace=True)
df = df[(df.Year<2020) & (df.Year>=2017)]

df_income = df.copy()
```
