## 2) Make a plot

At first we will see how to do a simple plot just to show you the general idea of plotting. Later I will show you different kinds of plots.

```python
#Define some dummy data
x = np.arange(10)
y = np.arange(10)
```

```python
#build the canvas
fig, ax = plt.subplots(nrows=1, ncols=1, figsize = (10,6))

#make the plot
ax.plot(x,y)
```

```python
#define the canvas
fig, axs = plt.subplots(nrows=2, ncols=2, figsize = (10,6))

#make the plot for each canva
#think at each canva as a separated unit

#-------------------------------------------
ax = axs[0,0]
ax.plot(x,y)
#-------------------------------------------
ax = axs[0,1]
ax.plot(x**2,y)
#-------------------------------------------
ax = axs[1,0]
ax.plot(x,y**2)
#-------------------------------------------
ax = axs[1,1]
ax.plot(x**2,y**2)
#-------------------------------------------
```

```python
#define the canvas
fig, axs = plt.subplots(nrows=2, ncols=2, figsize = (10,6))

#make the plot for each canva
#think at each canva as a separated unit

#-------------------------------------------
ax = axs[0,0]
ax.plot(x,y, color = 'red')
#-------------------------------------------
ax = axs[0,1]
ax.plot(x**2,y, color = 'green')
#-------------------------------------------
ax = axs[1,0]
ax.plot(x,y**2, color = 'blue')
#-------------------------------------------
ax = axs[1,1]
ax.plot(x**2,y**2, color = 'orange')
#-------------------------------------------
```

## 3) How to adjust the figure look

### Labels

```python
#define the canvas
fig, axs = plt.subplots(nrows=2, ncols=2, figsize = (10,6))

#make the plot and the labels for each canva

#-------------------------------------------
ax = axs[0,0]
ax.plot(x,y, color = 'red',linestyle = '-')
ax.plot(x**2,y, color = 'green',linestyle = '--')

ax.set_xlabel('we are plotting x**2')
ax.set_ylabel('we are plotting y')
ax.set_title('red and green')
#-------------------------------------------
ax = axs[0,1]
ax.plot(x**2,y, color = 'green',linestyle = '--')
ax.plot(x**2,y**2, color = 'orange',linestyle = ':')
ax.set_xlabel('we are plotting x**2')
ax.set_ylabel('we are plotting y**2')
ax.set_title('orange and green')
#-------------------------------------------
ax = axs[1,0]
ax.plot(x,y**2, color = 'blue',linestyle = '-.')
ax.plot(x,y, color = 'red',linestyle = '-')
ax.set_xlabel('we are plotting the third canva')
ax.set_ylabel('we are plotting...')
ax.set_title('red and blue')
#-------------------------------------------
ax = axs[1,1]
ax.plot(x**2,y**2, color = 'orange',linestyle = ':')
ax.plot(x,y**2, color = 'blue',linestyle = '-.')
ax.set_xlabel('fourth canva x')
ax.set_ylabel('fourth canva y')
ax.set_title('orange and blue')
#-------------------------------------------

fig.tight_layout()
```

### Fonts

```python
#define the canvas
fig, axs = plt.subplots(nrows=2, ncols=2, figsize = (15,10))

#make the plot and the labels for each canva

#-------------------------------------------
ax = axs[0,0]
ax.plot(x,y, color = 'red',linestyle = '-')
ax.plot(x**2,y, color = 'green',linestyle = '--')

ax.set_xlabel('we are plotting x**2', fontsize = 18)
ax.set_ylabel('we are plotting y', fontsize = 18)
ax.set_title('red and green', fontsize = 25)
#-------------------------------------------
ax = axs[0,1]
ax.plot(x**2,y, color = 'green',linestyle = '--')
ax.plot(x**2,y**2, color = 'orange',linestyle = ':')

ax.set_xlabel('we are plotting x**2', fontsize = 18)
ax.set_ylabel('we are plotting y**2', fontsize = 18)
ax.set_title('orange and green', fontsize = 25)
#-------------------------------------------
ax = axs[1,0]
ax.plot(x,y**2, color = 'blue',linestyle = '-.')
ax.plot(x,y, color = 'red',linestyle = '-')

ax.set_xlabel('we are plotting the third canva', fontsize = 18)
ax.set_ylabel('we are plotting...', fontsize = 18)
ax.set_title('red and blue', fontsize = 25)
#-------------------------------------------
ax = axs[1,1]
ax.plot(x**2,y**2, color = 'orange',linestyle = ':')
ax.plot(x,y**2, color = 'blue',linestyle = '-.')

ax.set_xlabel('fourth canva x', fontsize = 18)
ax.set_ylabel('fourth canva y', fontsize = 18)
ax.set_title('orange and blue', fontsize = 25)
#-------------------------------------------

fig.tight_layout()
```

```python
#define the canvas
fig, axs = plt.subplots(nrows=2, ncols=2, figsize = (15,10))

#make the plot and the labels for each canva

#-------------------------------------------
ax = axs[0,0]
ax.plot(x,y, color = 'red',linestyle = '-')
ax.plot(x**2,y, color = 'green',linestyle = '--')

ax.set_xlabel('we are plotting x**2', fontsize = 18)
ax.set_ylabel('we are plotting y', fontsize = 18)
ax.set_title('red and green', fontsize = 25)

ax.tick_params(axis='both', labelsize=15)
#-------------------------------------------
ax = axs[0,1]
ax.plot(x**2,y, color = 'green',linestyle = '--')
ax.plot(x**2,y**2, color = 'orange',linestyle = ':')
ax.set_xlabel('we are plotting x**2', fontsize = 18)
ax.set_ylabel('we are plotting y**2', fontsize = 18)
ax.set_title('orange and green', fontsize = 25)
#-------------------------------------------
ax = axs[1,0]
ax.plot(x,y**2, color = 'blue',linestyle = '-.')
ax.plot(x,y, color = 'red',linestyle = '-')
ax.set_xlabel('we are plotting the third canva', fontsize = 18)
ax.set_ylabel('we are plotting...', fontsize = 18)
ax.set_title('red and blue', fontsize = 25)
#-------------------------------------------
ax = axs[1,1]
ax.plot(x**2,y**2, color = 'orange',linestyle = ':')
ax.plot(x,y**2, color = 'blue',linestyle = '-.')
ax.set_xlabel('fourth canva x', fontsize = 18)
ax.set_ylabel('fourth canva y', fontsize = 18)
ax.set_title('orange and blue', fontsize = 25)

ax.tick_params(axis='both', labelsize=15)
#-------------------------------------------

fig.tight_layout()
```

### Legend

```python
x = np.arange(10)
y = np.arange(10)
```

```python
#define the canvas
fig, axs = plt.subplots(nrows=2, ncols=2, figsize = (15,10))

#make the plot and the labels for each canva

#-------------------------------------------
ax = axs[0,0]
ax.plot(x,y, color = 'red',linestyle = '-')
ax.plot(x**2,y, color = 'green',linestyle = '--')

ax.set_xlabel('we are plotting x**2', fontsize = 18)
ax.set_ylabel('we are plotting y', fontsize = 18)
ax.set_title('red and green', fontsize = 25)

ax.tick_params(axis='both', labelsize=15)

ax.legend(['firstline','secondline'])
#-------------------------------------------
ax = axs[0,1]
ax.plot(x**2,y, color = 'green',linestyle = '--')
ax.plot(x**2,y**2, color = 'orange',linestyle = ':')
ax.set_xlabel('we are plotting x**2', fontsize = 18)
ax.set_ylabel('we are plotting y**2', fontsize = 18)
ax.set_title('orange and green', fontsize = 25)
ax.tick_params(axis='both', labelsize=15)

ax.legend(['firstline','secondline'],fontsize = 14)
#-------------------------------------------
ax = axs[1,0]
ax.plot(x,y**2, color = 'blue',linestyle = '-.')
ax.plot(x,y, color = 'red',linestyle = '-')
ax.set_xlabel('we are plotting the third canva', fontsize = 18)
ax.set_ylabel('we are plotting...', fontsize = 18)
ax.set_title('red and blue', fontsize = 25)
ax.tick_params(axis='both', labelsize=15)

ax.legend(['firstline','secondline'],fontsize = 14, loc ='lower right')
#-------------------------------------------
ax = axs[1,1]
ax.plot(x**2,y**2, color = 'orange',linestyle = ':')
ax.plot(x,y**2, color = 'blue',linestyle = '-.')
ax.set_xlabel('fourth canva x', fontsize = 18)
ax.set_ylabel('fourth canva y', fontsize = 18)
ax.set_title('orange and blue', fontsize = 25)
ax.tick_params(axis='both', labelsize=15)

ax.legend(['firstline','secondline'],fontsize = 14, loc ='upper left')
#-------------------------------------------

fig.tight_layout()
```

### Scatterplot

```python
x = np.random.random(10)
y = np.random.random(10)
```

```python
#define the canvas
fig, axs = plt.subplots(nrows=2, ncols=2, figsize = (15,10))

#make the plot and the labels for each canva

#-------------------------------------------
ax = axs[0,0]
ax.scatter(x,y)

ax.set_xlabel('Random number x', fontsize = 18)
ax.set_ylabel('Random number y', fontsize = 18)
ax.set_title('Simple scatterplot', fontsize = 25)

ax.tick_params(axis='both', labelsize=15)
#-------------------------------------------
ax = axs[0,1]
ax.scatter(x,y, color = 'red', s = 200)

ax.set_xlabel('Random number x', fontsize = 18)
ax.set_ylabel('Random number y', fontsize = 18)
ax.set_title('Fixed color and size', fontsize = 25)

ax.tick_params(axis='both', labelsize=15)
#-------------------------------------------
ax = axs[1,0]

sizes = np.random.random(10)*1000
ax.scatter(x,y, color = 'red', s = sizes)


ax.set_xlabel('Random number x', fontsize = 18)
ax.set_ylabel('Random number y', fontsize = 18)
ax.set_title('Fixed color and variable size', fontsize = 25)

ax.tick_params(axis='both', labelsize=15)
#-------------------------------------------
ax = axs[1,1]

sizes = np.random.random(10)*1000
colors = np.random.random(10)
ax.scatter(x,y, s = sizes, c=colors, cmap = 'Blues')

ax.set_xlabel('Random number x', fontsize = 18)
ax.set_ylabel('Random number y', fontsize = 18)
ax.set_title('Variable color and variable size', fontsize = 25)



ax.tick_params(axis='both', labelsize=15)
#-------------------------------------------

fig.tight_layout()
```

Link for colormap: https://matplotlib.org/3.5.0/tutorials/colors/colormaps.html

### Histogram

```python
distribution = np.random.randn(10000)


fig, axs = plt.subplots(nrows=2, ncols=2, figsize = (20,16))

ax = axs[0,0]
ax.hist(distribution)

ax.set_xlabel('Random number', fontsize = 18)
ax.set_ylabel('Counts', fontsize = 18)
ax.set_title('Normal distribution', fontsize = 25)

ax.tick_params(axis='both', labelsize=15)



ax = axs[0,1]
n_bins = 1000
ax.hist(distribution,n_bins)

ax.set_xlabel('Random number', fontsize = 18)
ax.set_ylabel('Counts', fontsize = 18)
ax.set_title('Normal distribution', fontsize = 25)

ax.tick_params(axis='both', labelsize=15)

ax = axs[1,0]
n_bins = 100
ax.hist(distribution,n_bins,density = True)

ax.set_xlabel('Random number', fontsize = 18)
ax.set_ylabel('Density (=counts_per_bin/total_counts)', fontsize = 18)
ax.set_title('Normal distribution', fontsize = 25)

ax.tick_params(axis='both', labelsize=15)

ax = axs[1,1]
n_bins = 100
ax.hist(distribution,n_bins,density = True,histtype='step',color='red')

ax.set_xlabel('Random number', fontsize = 18)
ax.set_ylabel('Density (=counts_per_bin/total_counts)', fontsize = 18)
ax.set_title('Normal distribution', fontsize = 25)

ax.tick_params(axis='both', labelsize=15)
```

### Boxplot

```python
distribution = np.random.rand(10000)


fig, axs = plt.subplots(nrows=1, ncols=3, figsize = (20,6))

ax = axs[0]
ax.boxplot(distribution)

ax.set_xlabel('Label', fontsize = 20)
ax.set_ylabel('Random number distribution', fontsize = 20)
ax.set_title('Boxplot anatomy', fontsize = 25)

ax.tick_params(axis='both', labelsize=15)


ax = axs[1]
continous_distro = np.random.rand(10000)
normal_distro = np.random.randn(10000)
labels =['Continous','Normal']
ax.boxplot([continous_distro,normal_distro],labels = labels)

ax.set_xlabel('Type of distribution', fontsize = 20)
ax.set_ylabel('Distribution', fontsize = 20)
ax.set_title('Distribution comparison with boxplot', fontsize = 25)

ax.tick_params(axis='both', labelsize=18)


ax = axs[2]

normal_distro_1 = np.random.randn(10000)
normal_distro_2 = 3*np.random.randn(10000)+3
normal_distro_3 = 0.5*np.random.randn(10000)-3

labels =['Normal1','Normal2','Normal3']
ax.boxplot([normal_distro_1, normal_distro_2, normal_distro_3],showfliers=False,labels = labels)

ax.set_xlabel('Type of distribution', fontsize = 20)
ax.set_ylabel('Distribution', fontsize = 20)
ax.set_title('Distribution comparison with boxplot', fontsize = 25)

ax.tick_params(axis='both', labelsize=18)

fig.tight_layout()
```

### Violinplot

```python
distribution = np.random.rand(10000)


fig, axs = plt.subplots(nrows=1, ncols=3, figsize = (20,6))

ax = axs[0]
ax.violinplot(distribution)

ax.set_xlabel('Labels', fontsize = 20)
ax.set_ylabel('Distribution', fontsize = 20)
ax.set_title('violinplot anatomy', fontsize = 25)

ax.tick_params(axis='both', labelsize=15)


ax = axs[1]
continous_distro = np.random.rand(10000)
normal_distro = np.random.randn(10000)
ax.violinplot([continous_distro,normal_distro])

ax.set_xlabel('Labels', fontsize = 20)
ax.set_ylabel('Distribution', fontsize = 20)
ax.set_title('Distribution comparison', fontsize = 25)

ax.tick_params(axis='both', labelsize=18)


ax = axs[2]

normal_distro_1 = np.random.randn(10000)
normal_distro_2 = 3*np.random.randn(10000)+3
normal_distro_3 = 0.5*np.random.randn(10000)-3

ax.violinplot([normal_distro_1, normal_distro_2, normal_distro_3])

ax.set_xlabel('Labels', fontsize = 20)
ax.set_ylabel('Distribution', fontsize = 20)
ax.set_title('Distribution comparison', fontsize = 25)

ax.tick_params(axis='both', labelsize=18)

fig.tight_layout()
```

### Ticks range

```python
x = np.random.random(100)
y = np.random.random(100)

sizes = np.random.random(100)*100
colors = np.random.random(100)
```

```python
#define the canvas
fig, axs = plt.subplots(nrows=2, ncols=2, figsize = (15,10))

#make the plot and the labels for each canva

#-------------------------------------------
ax = axs[0,0]
ax.scatter(x,y, s = sizes, c=colors, cmap = 'Blues')

ax.set_xlabel('Random number x', fontsize = 18)
ax.set_ylabel('Random number y', fontsize = 18)
ax.set_title('Baseline', fontsize = 25)



ax.tick_params(axis='both', labelsize=15)
#-------------------------------------------
ax = axs[0,1]
ax.scatter(x,y, s = sizes, c=colors, cmap = 'Reds')

ax.set_xlabel('Random number x', fontsize = 18)
ax.set_ylabel('Random number y', fontsize = 18)
ax.set_title('Variable color and variable size', fontsize = 25)

ax.set(xlim=(0., 0.5),ylim=(0, 0.5))

ax.tick_params(axis='both', labelsize=15)
#-------------------------------------------
ax = axs[1,0]
ax.scatter(x,y, s = sizes, c=colors, cmap = 'seismic')

ax.set_xlabel('Random number x', fontsize = 18)
ax.set_ylabel('Random number y', fontsize = 18)
ax.set_title('Variable color and variable size', fontsize = 25)

ax.set(xlim=(0.25, 0.75),ylim=(0.1, 0.68))

ax.tick_params(axis='both', labelsize=15)
#-------------------------------------------
ax = axs[1,1]
ax.scatter(x,y, s = sizes, c=colors, cmap = 'cool')

ax.set_xlabel('Random number x', fontsize = 18)
ax.set_ylabel('Random number y', fontsize = 18)
ax.set_title('Variable color and variable size', fontsize = 25)

ax.set(xlim=(0.35, 0.9),ylim=(0.2, 0.65))

ax.tick_params(axis='both', labelsize=15)
#-------------------------------------------

fig.tight_layout()
```

### Colorbar

```python
x = np.random.random(100)
y = np.random.random(100)

sizes = np.random.random(100)*100
colors = np.random.random(100)
```

```python
#define the canvas
fig, axs = plt.subplots(nrows=2, ncols=2, figsize = (18,10))

#make the plot and the labels for each canva

#-------------------------------------------
ax = axs[0,0]
plot = ax.scatter(x,y, s = sizes, c=colors, cmap = 'Blues')
fig.colorbar(plot, ax=ax)

ax.set_xlabel('Random number x', fontsize = 18)
ax.set_ylabel('Random number y', fontsize = 18)
ax.set_title('Baseline', fontsize = 25)


ax.tick_params(axis='both', labelsize=15)
#-------------------------------------------
ax = axs[0,1]
plot = ax.scatter(x,y, s = sizes, c=colors, cmap = 'Reds')
fig.colorbar(plot, ax=ax, shrink=0.7)

ax.set_xlabel('Random number x', fontsize = 18)
ax.set_ylabel('Random number y', fontsize = 18)
ax.set_title('Variable color and variable size', fontsize = 25)

ax.set(xlim=(0., 0.5),ylim=(0, 0.5))

ax.tick_params(axis='both', labelsize=15)
#-------------------------------------------
ax = axs[1,0]
plot = ax.scatter(x,y, s = sizes, c=colors, cmap = 'seismic')
cb = fig.colorbar(plot, ax=ax, shrink=0.7)
cb.set_label('Colors',rotation=270,labelpad = 20, fontsize = 18)


ax.set_xlabel('Random number x', fontsize = 18)
ax.set_ylabel('Random number y', fontsize = 18)
ax.set_title('Variable color and variable size', fontsize = 25)

ax.set(xlim=(0.25, 0.75),ylim=(0.1, 0.68))

ax.tick_params(axis='both', labelsize=15)
#-------------------------------------------
ax = axs[1,1]
plot = ax.scatter(x,y, s = sizes, c=colors, cmap = 'cool')
cb = fig.colorbar(plot, ax=ax, shrink=0.7)
cb.set_label('Colors',labelpad = 20, fontsize = 18)
cb.ax.tick_params(labelsize=15)

ax.set_xlabel('Random number x', fontsize = 18)
ax.set_ylabel('Random number y', fontsize = 18)
ax.set_title('Variable color and variable size', fontsize = 25)

ax.set(xlim=(0.35, 0.9),ylim=(0.2, 0.65))

ax.tick_params(axis='both', labelsize=15)
#-------------------------------------------

fig.tight_layout()
```
