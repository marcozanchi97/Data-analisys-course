The pandas library is the most important tool for analyzing data with Python. Pandas allows you to clean, transform and analyzie your data. 

Data are genrally stored in csv files (similar to Excel files). Pandas will extract the data from that csv fill into a DataFrame — a table, basically — then allows you to calculate statistics and answer questions about the data in  very simple and clean way. Some examples of questions and operation allowed by Pandas:
- What's the average, median, max, or min of each column?
- Does column A correlate with column B?
- Clean the data by doing things like removing missing values and filtering rows or columns by some criteria
- Store the cleaned, transformed data back into a CSV, other file or database

## Clean the dataset

The hardest thing in data analysis is to clean and re-organize data

### 1) Drop duplicates

```python
df = df.drop_duplicates()
df
```

```python
df.drop_duplicates(inplace = True)
```

An example:

```python
df_duplicated = df.append(df)
df_duplicated
```

```python
df_duplicated.drop_duplicates()
```

### 2) Change column names

```python
df.rename(columns={'Score': 'SCORE', 'Type': 'TYPE'}, inplace=True)

df
```

```python
#TIP: how to automatically change the column names to lowercase
df.columns = [col.lower() for col in df]
df
```

### 3) Missing or wrong values

```python
bad_col = []
for i in range(len(df)):
    bad_col.append(np.nan)
df['messy column'] = bad_col
```

```python
#check for null values: NaN

df.isnull()
```

```python
#check for null values in a compact way
df.isnull().sum()
```

```python
#get rid of null values
df.dropna()
```

```python
#get rid of null values
df.dropna(axis = 1, inplace = True)
df
```

### 4) Drop columns

```python
df.drop(labels=['air date'],axis = 1,inplace = True)
df
```

### 3) Make a subset of your df

```python
#by columns name
df_sub = df[['name','score']]
df_sub.head(5)
```

```python
#by index
df_sub = df.iloc[0:1000]
df_sub
```

```python
#by conditional selections

#df_sub = df[condition]

df_sub = df[df['num. of episodes']>100]
df_sub
```

```python
#by conditional selections
df_sub = df[df['studio']=="['Sunrise']"]
df_sub
```

```python
#by conditional selections
df_sub = df[df['demographic']=="Shounen"]
df_sub
```

```python
#by conditional selections

#df_sub = df[(condition1)&(condition2)&(condtion3)]
df_sub = df[(df['score']>9)&(df['demographic']=="Shounen")]
df_sub
```
